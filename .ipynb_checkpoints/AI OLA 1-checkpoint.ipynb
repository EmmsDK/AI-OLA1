### Task 1: Data Exploration and Cleaning

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import spearmanr

df = pd.read_csv("data_science_job.csv")
````

````python
df.head()
````

````python
df.describe()
````

````python
# Lists each column type. Useful for letting us know how we can interpolate the different data.
# Numerical data can be interpolated with df.interpolate().
# Categorical data, object = string, can use the df.mode() to replace missing data with the most used value in the column.
df.info()
````

````python
# Shows count of missing values per column
print(df.isnull().sum())
````

````python
gender_counts = df['gender'].value_counts(dropna=False)
print(gender_counts)
````

````python
df = df.drop("major_discipline", axis=1)
````

````python
print(df.isnull().sum())
````

````python
# dropna(subset=['Gender']) removes NaN values.
# .query("Gender != 'Other'") removes "Other" values.
df = df.dropna(subset=['gender']).query("gender != 'Other'")

# This line of code drops beoith company_size and company_type
df = df.drop(["company_size", "company_type"], axis=1)

gender_counts = df['gender'].value_counts(dropna=False)
print(gender_counts)
````

### Preparing data for plotting. Gender is changed to either 1 or 0 (male/female) as is relevant experience (1 = relevant experience | 0 = no relevant experience)
### String columns with a NaN value is replaced with the most common value in the column

````python
# Interpolates missing values of the type number. It calculates a fitting value within the current values and then adds it to a missing cell

# Althought this fixes all NaN numbers, it does not provide an accurate answer based on the whole column, it only looks to it's neightbours
# To provide a true accurate number we should calculate the mean of each column and fill the NaN numbers out with the mean
df[df.select_dtypes(include=['number']).columns] = df.select_dtypes(include=['number']).interpolate(method='linear')

# Gender is either male or female we can replace the values with 1 for male and 0 for female. The lambda expression does that.
# Same logic ais applied to relevant experience
df['gender'] = df['gender'].apply(lambda x: 1 if x == 'Male' else 0)
df['relevent_experience'] = df['relevent_experience'].apply(lambda x: True if x == 'Has relevent experience' else False)

# replaces a missing value with the most common value in the column.
df['enrolled_university'].fillna(df['enrolled_university'].mode()[0], inplace=True)
df['education_level'].fillna(df['education_level'].mode()[0], inplace=True)

df.head()

gender_counts = df['gender'].value_counts(dropna=False)
print(gender_counts)
print(df.isnull().sum())
````

````python
df_male = df[df['gender']==1]

plt.figure(figsize=(10, 6))
sns.histplot(data=df_male, x='experience', kde=True, color="#f56f42")
plt.title('Experience Distribution of Male Employees')
plt.xlabel('Years of Experience')
plt.ylabel('Count')
plt.show()
````

````python
df_female = df[df['gender'] == 0]

plt.figure(figsize=(10, 6))
sns.histplot(data=df_female, x='experience', kde=True)
plt.title('Experience Distribution of Female Employees')
plt.xlabel('Years of Experience')
plt.ylabel('Count')
plt.show()
````

### Task 2: Feature Engineering and Descriptive Statistics

````python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='training_hours', kde=True, hue='gender', multiple='dodge')
plt.title('training_hours by gender')
plt.show()
````

````python
#Feature Engineering

# Define bins and labels for city_development_index
bins = [0.0, 0.6, 0.8, 1.0]  # Bin edges
labels = ["Low Development (≤0.6)", "Medium Development (0.6 - 0.8)", "High Development (0.8 - 1.0)"]

# Create a new column with categories
df["development_category"] = pd.cut(df["city_development_index"], bins=bins, labels=labels, include_lowest=True)

# Force gender mapping to replace 0 -> Female, 1 -> Male
df["gender"] = df["gender"].replace({0: "Female", 1: "Male"})

# Verify mapping applied correctly
print("Unique gender values:", df["gender"].unique())

# Group by development category and gender
gender_distribution = df.groupby("development_category", observed=False)["gender"].value_counts().unstack()

# Plot pie charts for each development category
for category in gender_distribution.index:
    plt.figure(figsize=(6, 6))
    gender_distribution.loc[category].plot(kind="pie", autopct="%1.1f%%", startangle=90)
    plt.title(f"Gender Distribution in {category}")
    plt.ylabel("")  # Remove y-axis label
    plt.show()
````

````python
### DESCRIPTIVE STATISTICS ###

### IDENTIFY AND REMOVE OUTLIERS USING IQR METHOD ###

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df["training_hours"].quantile(0.25)
Q3 = df["training_hours"].quantile(0.75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Display the calculated bounds
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# Filter the DataFrame to remove outliers
df_th_clear = df[(df["training_hours"] >= lower_bound) & (df["training_hours"] <= upper_bound)]

# Display the number of rows removed
rows_removed = len(df) - len(df_th_clear)
print(f"Number of rows removed due to outliers: {rows_removed}")

# Numerical Features: Mean, Median, Standard Deviation
num_features = ["training_hours"]  # Add more numerical features if needed
for col in num_features:
    print(f"\nStatistics for {col}:")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Standard Deviation: {df[col].std():.2f}")

# Categorical Features: Frequency Count
cat_features = ["education_level"]  # Add more categorical features if needed
for col in cat_features:
    print(f"\nCategory Counts for {col}:")
    print(df[col].value_counts())

### VISUALIZATION ###

## BOX PLOT FOR NUMERICAL FEATURES (OUTLIERS) ##
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_th_clear["training_hours"], color="orange")
plt.title("Box Plot of Training Hours")
plt.xlabel("Training Hours")
plt.show()

# BAR PLOT FOR CATEGORICAL FEATURE: COMPANY TYPE
plt.figure(figsize=(10, 6))

# Get value counts for company type
education_level = df["education_level"].value_counts()

# Corrected bar plot with explicit hue and legend=False
sns.barplot(x=education_level.index, y=education_level.values, hue=education_level.index, palette="viridis", dodge=False)

plt.title("Distribution of Education Level")
plt.xlabel("Education Level")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate labels for readability
plt.show()
````

````python
# Correlation Test
````

````python
# Convert categorical education levels into numerical codes
df['education_level_encoded'] = df['education_level'].astype('category').cat.codes

correlation = df[['education_level_encoded', 'training_hours']].corr()

print("Correlation matrix:")
print(correlation)
````

````python
# Compute Spearman's correlation
spearman_corr, p_value = spearmanr(df['education_level_encoded'], df['training_hours'])

print(f"Spearman correlation: {spearman_corr:.3f}")
print(f"P-value: {p_value:.3f}")
````

### Interpreting Results:
* Correlation Coefficient (r):
  * +1.0 → Strong positive correlation (higher education = more training hours)
  * 0.0 → No correlation
  * -1.0 → Strong negative correlation (higher education = fewer training hours)
* P-value (p):
  * If p < 0.05, the correlation is statistically significant.

### Insights:
* There is no strong relationship between education level and training hours.
  * The Spearman correlation is -0.013, which is very close to 0, indicating no meaningful trend.
  * This means that higher education does not necessarily lead to more or fewer training hours.
* The result is not statistically significant.
  * The p-value (0.106) is greater than 0.05, meaning we fail to reject the null hypothesis.
  * This suggests that any slight variation in training hours based on education level is likely due to random chance rather than a real pattern.
* Training hours might be influenced by other factors rather than education.
  * Since education level does not strongly correlate with training hours, other variables like job role, industry, company policies, or experience level might have a bigger impact.
  * Further analysis could explore these other factors to understand what actually determines training hours.
